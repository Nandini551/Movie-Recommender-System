```python
#Convert function
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):#To convert string of list to list 
        L.append(i['name'])
    return L
```

```python
#combine all the genres
movies['genres']=movies['genres'].apply(convert)
```

```python
#combine all the keywords
movies['keywords']=movies['keywords'].apply(convert)
```

```python
#convert3 function
def convert3(obj):
    L=[]
    count=0
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count+=1
        else:
            break
    return L
```

```python
#Taking first 3 members of the cast
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#Fetching only the director from the crew
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
#TEXT VECTORIZATION: To compare any two paragraphs used for recommending
#We use bag of words technique
#We take 5000 most common words from all the tags
#Then we calculate each word's frequency
#We remove stopwords,apply stemming
#Calculate cosine distance between any 2 vectors
#Obtain similarities and sort them to use them for recommendations
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
#making an array
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
#sparse matrix
vectors[0]
```

```python
cv.get_feature_names()
```

```python
#Calculating Cosine distance instead of Eucledian distance between 2 vectors beacause it fails for large amount of data
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0].shape
```

```python
similarity[0]
```

```python
similarity[1]
```

```python
#sort the similarities
sorted(similarity[0],reverse=True)
```

```python
list(enumerate(similarity[0]))
```

```python
movie_list=sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
movie_list
```

```python
#Recommender function
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```
